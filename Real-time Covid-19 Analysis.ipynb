Query1 - News has Reported that, Covid-19is more likely to affect elder population as compared to the youth, is that true?

```python
def plot_pie_charts(x, y, title):
    c = random.choices(list(mcolors.CSS4_COLORS.values()),k = 10)
    plt.figure(figsize=(20,15))
    plt.title(title, size=20)
    plt.pie(y, colors = c)
    plt.legend(x, loc='best', fontsize=15)
    plt.show()
```

```python
plot_pie_charts(df1['AgeGroup'], df1['TotalCases'], 'Age distribution')
```

Observation : We can see that most of the cases are of people 20-29 age range with the ranges 30-39,40-49,50-59,60-69 also having a high number of cases.

Query 2 - The Prime Minister has asked you to plan the migration plans of the labour's as it has become a huge problem, we want them to get the facilities but we have limited resources in cities, how will you handle this situation?

```python
def plot_bar_graphs(x, y, title):
    plt.figure(figsize=(20, 12))
    plt.barh(x, y)
    plt.title(title, size=20)
    plt.xticks(size=15)
    plt.yticks(size=20)
    plt.show()
```

```python
plot_bar_graphs(df2['State/UT'], df2['NumRuralHospitals_NHP18'], '# of Rural Hospitals in different regions')
```

Observation : 
●	Uttar pradesh has the most and is the only state with over 4000 rural hospitals.
●	Next followed by Karnataka with 2500 rural hospital and stands at second place.
